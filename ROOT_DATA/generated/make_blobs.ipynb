```python
import sklearn.datasets as datagen
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D # 3d plotting
import matplotlib.pyplot as plt
import pandas as pd
```

```python
colors = {0:'red', 1:'orange', 2:'green', 3:'blue'}
```

```python
# {{{datagen_blobs_2dimg}}}

# {{{datagen_blobs_2dcode
X, y = datagen.make_blobs(centers=4, n_samples=100, n_features=2,cluster_std=1.0,
                          center_box=(-10, 10),
                          random_state=42, shuffle=True)
# END}}}
df = pd.DataFrame(dict(feature_0=X[:,0], feature_1=X[:,1], y=y))
groups = df.groupby('y')

# figure
fig, ax = plt.subplots()
for key, group in groups:
    group.plot(ax=ax, kind='scatter', x='feature_0', y='feature_1', label=key, color=colors[key])
plt.show()
```

```python
# {{{datagen_blobs_3dimg}}}

# {{{datagen_blobs_3dcode
X, y = datagen.make_blobs(centers=4, n_samples=100, n_features=3, random_state=42)
# END}}}
df = pd.DataFrame(dict(feature_0=X[:,0], feature_1=X[:,1], feature_2=X[:,2], y=y))
groups = df.groupby('y')

# figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for key, group in groups:
    ax.scatter3D(xs=group['feature_0'], ys=group['feature_1'], zs=group['feature_2'], 
                 label=key, c=colors[key])
plt.show()
```
