```python
import sklearn.datasets as datagen
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D # 3d plotting
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
colors = {0:'blue', 1:'orange', 2:'green', 3:'red'}
```

```python
# {{{kernelized_2class4clust_2dimg}}}

X, y = datagen.make_blobs(centers=4, n_samples=100, n_features=2,cluster_std=1,
                          center_box=(-12, 12),
                          random_state=18, shuffle=True)
y = y%2 # make only two labels, but still w/4 clusters


df = pd.DataFrame(dict(feature_0=X[:,0], feature_1=X[:,1], y=y))
groups = df.groupby('y')

fig, ax = plt.subplots()
for key, group in groups:
    group.plot(ax=ax, kind='scatter', x='feature_0', y='feature_1', label=key, color=colors[key])
plt.show()
```

```python
# {{{kernelized_2class4clust_3dimg}}}

# figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for key, group in groups:
    ax.scatter3D(xs=group['feature_0'], 
                 ys=group['feature_1'], 
                 zs=(group['feature_1']**2), 
                 label=key, c=colors[key])
    ax.set_xlabel("feature_0")
    ax.set_ylabel("feature_1")
    ax.set_zlabel("feature_1**2")
    ax.view_init(elev=8., azim=8)
plt.show()
```
